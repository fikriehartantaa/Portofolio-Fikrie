# Data Understanding

Berikut adalah arti dari masing-masing kolom:

1. **user_id**: ID unik untuk setiap pengguna.  
2. **location**: Lokasi pengguna (biasanya berupa gabungan kota, negara bagian, dan negara).  
3. **age**: Usia pengguna dalam satuan tahun.  
4. **isbn**: Nomor ISBN (International Standard Book Number) untuk identifikasi unik buku.  
5. **rating**: Rating atau ulasan yang diberikan oleh pengguna terhadap buku (biasanya dalam skala tertentu, misalnya 1-5).  
6. **book_title**: Judul buku.  
7. **book_author**: Nama penulis buku.  
8. **year_of_publication**: Tahun penerbitan buku.  
9. **publisher**: Nama penerbit buku.  
10. **img_s**: URL gambar kecil dari sampul buku.  
11. **img_m**: URL gambar medium dari sampul buku.  
12. **img_l**: URL gambar besar dari sampul buku.  
13. **Summary**: Ringkasan isi buku.  
14. **Language**: Bahasa yang digunakan dalam buku.  
15. **Category**: Kategori atau genre buku (misalnya, fiksi, non-fiksi).  
16. **city**: Kota asal pengguna.  
17. **state**: Negara bagian asal pengguna (jika berlaku).  
18. **country**: Negara asal pengguna.

# *1. Collaborative Filtering*

## Memilih Kolom

```python
df_cf = df[['user_id', 'book_title', 'rating']].dropna()
```

## Filter Buku dan Pengguna

```python
book_counts = df_cf['book_title'].value_counts()
active_books = book_counts[book_counts >= 10].index
df_cf = df_cf[df_cf['book_title'].isin(active_books)]

user_counts = df_cf['user_id'].value_counts()
active_users = user_counts[user_counts >= 5].index
df_cf = df_cf[df_cf['user_id'].isin(active_users)]
```

## Reset Index

```python
df_cf.reset_index(drop=True, inplace=True)
```

#  Implementasi Collaborative Filtering

```python
# !pip install scikit-surprise
```

```python
from surprise import Dataset, Reader

reader = Reader(rating_scale=(df_cf['rating'].min(), df_cf['rating'].max()))
data = Dataset.load_from_df(df_cf[['user_id', 'book_title', 'rating']], reader)
```

## Model SVD

```python
from surprise import SVD
from surprise.model_selection import cross_validate

# Inisialisasi model SVD
svd = SVD()

# Evaluasi menggunakan cross-validation
cv_results = cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

## Prediksi dan Rekomendasi

```python
# Fit model pada dataset penuh
trainset = data.build_full_trainset()
svd.fit(trainset)

# Fungsi rekomendasi
```

## Fungsi Rekomendasi

```python
def recommend_cf(user_id, top_n=10):
    user_rated_books = df_cf[df_cf['user_id'] == user_id]['book_title'].unique()
    all_books = df_cf['book_title'].unique()
    books_to_predict = [book for book in all_books if book not in user_rated_books]

    predictions = [(book, svd.predict(user_id, book).est) for book in books_to_predict]
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:top_n]
    return [rec[0] for rec in recommendations]
```

# *2. Content-Based Filtering*

## Memilih Kolom

```python
df_cb = df[['book_title', 'Summary', 'book_author', 'Category', 'Language']].drop_duplicates(subset='book_title')
```

## Handling Missing Value

```python
df_cb['Summary'].fillna('', inplace=True)
df_cb['Category'].fillna('Unknown', inplace=True)
df_cb['Language'].fillna('Unknown', inplace=True)
```

## Preprocess Teks

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

df_cb['Summary_cleaned'] = df_cb['Summary'].apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df_cb['Summary_cleaned'])
```

## Encode Atribut Kategorikal

```python
print(df_cb['Category'].head())
print(df_cb['Category'].apply(type).value_counts())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

# Tangani list kosong di kolom Category
df_cb['Category'] = df_cb['Category'].apply(lambda x: ['Unknown'] if len(x) == 0 else x)

# Encode Category dengan MultiLabelBinarizer
mlb = MultiLabelBinarizer()
category_encoded = mlb.fit_transform(df_cb['Category'])

# Encode Language dengan pd.get_dummies
language_encoded = pd.get_dummies(df_cb['Language']).values

# Hasil Encoded
print("Category Encoded Shape:", category_encoded.shape)
print("Language Encoded Shape:", language_encoded.shape)
```

## Gabungkan Fitur

```python
from scipy.sparse import hstack

content_features = hstack([tfidf_matrix, category_encoded, language_encoded])
```

# Implementasi Content-Based Filtering

## Hitung Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

book_similarity = cosine_similarity(content_features)
book_indices = pd.Series(df_cb.index, index=df_cb['book_title']).drop_duplicates()
```

## Fungsi Rekomendasi

```python
def recommend_cb(book_title, top_n=10):
    book_idx = book_indices[book_title]
    similar_indices = book_similarity[book_idx].argsort()[::-1][1:top_n+1]
    return df_cb.iloc[similar_indices]['book_title'].tolist()
```

# *3. Hybrid Collaborative dan Content-Based*

```python
def hybrid_recommend(user_id, book_title, top_n=10, alpha=0.5):
    # Collaborative Filtering recommendations
    cf_recs = recommend_cf(user_id, top_n=top_n)

    # Content-Based Filtering recommendations
    cb_recs = recommend_cb(book_title, top_n=top_n)

    # Kombinasikan skor
    combined_scores = {}
    for idx, book in enumerate(cf_recs):
        combined_scores[book] = alpha * (top_n - idx)
    for idx, book in enumerate(cb_recs):
        combined_scores[book] = combined_scores.get(book, 0) + (1 - alpha) * (top_n - idx)

    # Urutkan rekomendasi
    hybrid_recs = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)
    return [rec[0] for rec in hybrid_recs[:top_n]]
```

# *4. Penggunaan(output) Tiap Algoritma*

## **4.1 Collaborative Filtering**

```python
# Masukkan user_id untuk mendapatkan rekomendasi
user_id = 12345  # Contoh user_id dalam dataset Anda
recommendations_cf = recommend_cf(user_id, top_n=10)
n = 0
# Tampilkan hasil rekomendasi
print("Rekomendasi berdasarkan Collaborative Filtering:")
for i in recommendations_cf:
  n += 1
  print(f'{n}. {i}')

```

## **4.2 Content-Based Filtering**

```python
# Masukkan book_title untuk mendapatkan buku serupa
book_title = "Classical Mythology"
recommendations_cb = recommend_cb(book_title, top_n=10)
n = 0
# Tampilkan hasil rekomendasi
print("Rekomendasi berdasarkan Content-Based Filtering:")
for i in recommendations_cb:
  n += 1
  print(f'{n}. {i}')
```

## **4.3 Hybrid Filtering**

```python
# Masukkan user_id dan book_title untuk kombinasi rekomendasi
user_id = 12345
book_title = "Classical Mythology"
recommendations_hybrid = hybrid_recommend(user_id, book_title, top_n=10, alpha=0.5)
n = 0
# Tampilkan hasil rekomendasi
print("Rekomendasi berdasarkan Hybrid Filtering:")
for i in recommendations_hybrid:
  n += 1
  print(f'{n}. {i}')
```
