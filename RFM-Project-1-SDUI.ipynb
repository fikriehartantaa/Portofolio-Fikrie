RFM Analysis adalah teknik pemasaran yang digunakan untuk menganalisis perilaku pelanggan berdasarkan tiga variabel utama: Recency, Frequency, dan Monetary value. 


1. **Recency (R)** 
    
    Mengukur seberapa baru transaksi terakhir dilakukan oleh pelanggan. Konsep ini didasarkan pada asumsi bahwa pelanggan yang baru saja melakukan pembelian lebih mungkin untuk membeli lagi dibandingkan dengan pelanggan yang sudah lama tidak bertransaksi.

2. **Frequency (F)**
    
    Mengukur seberapa sering pelanggan melakukan pembelian dalam periode waktu tertentu. Pelanggan yang lebih sering bertransaksi dianggap lebih mungkin untuk melakukan pembelian di masa depan.

3. **Monetary (M)**
    
    Mengukur total uang yang dihabiskan oleh pelanggan selama periode tertentu. Pelanggan yang menghabiskan lebih banyak uang dianggap lebih berharga bagi bisnis.


Proses RFM Analysis melibatkan pengurutan pelanggan berdasarkan masing-masing faktor RFM secara terpisah dan kemudian mengelompokkan mereka ke dalam kuintil (lima kelompok yang sama besar). Setiap kuintil diberi skor dari 1 hingga 5, dengan 5 menunjukkan nilai tertinggi. Skor ini kemudian digabungkan untuk memberikan gambaran lengkap tentang nilai pelanggan.

Untuk menentukan skor dengan skala 1 sampai 5 untuk setiap komponen RFM (Recency, Frequency, Monetary), Anda dapat menggunakan metode kuintil atau persentil. Berikut adalah langkah-langkahnya:

1. **Sortir Data**:
   - Urutkan data untuk setiap atribut RFM (Recency, Frequency, Monetary) dalam urutan menaik atau menurun.

2. **Pembagian ke dalam Kuintil**:
   - Bagi setiap atribut RFM menjadi lima bagian yang sama, di mana setiap bagian mewakili 20% dari total data. Ini dikenal sebagai pembagian kuintil.

3. **Penetapan Skor**:
   - Tetapkan skor 1 hingga 5 untuk setiap bagian, di mana skor 5 diberikan kepada kuintil dengan kontribusi tertinggi (misalnya, untuk Recency, skor 5 diberikan kepada pelanggan yang paling baru bertransaksi) dan skor 1 untuk kontribusi terendah.

Dengan cara ini, setiap pelanggan akan mendapatkan skor R, F, dan M yang masing-masing berkisar antara 1 hingga 5, yang kemudian dapat digunakan untuk segmentasi dan analisis lebih lanjut.

RFM Analysis berguna untuk meningkatkan segmentasi pelanggan dengan membagi mereka ke dalam berbagai kelompok untuk layanan personalisasi di masa depan dan untuk mengidentifikasi pelanggan yang lebih mungkin merespons promosi. Integrasi analisis RFM dengan teknik data mining dapat memberikan informasi yang berguna untuk pengambilan keputusan pemasaran.

Berikut adalah langkah-langkah detail untuk mengimplementasikan RFM Analysis:

1. **Pengumpulan Data**: 
   - Kumpulkan data transaksi pelanggan yang mencakup tanggal transaksi, frekuensi pembelian, dan jumlah uang yang dihabiskan. Data ini biasanya diambil dari sistem manajemen hubungan pelanggan (CRM) atau basis data penjualan.

2. **Pra-pemrosesan Data**:
   - Lakukan pra-pemrosesan data untuk memastikan kualitas data yang baik. Ini melibatkan penghapusan atribut yang tidak diperlukan, penanganan nilai yang hilang, dan normalisasi data. Misalnya, atribut seperti alamat email atau nomor telepon mungkin tidak relevan untuk analisis dan dapat dihapus.

3. **Penghitungan Skor RFM**:
   - Hitung skor untuk setiap komponen RFM:
     - **Recency**: Tentukan berapa lama sejak transaksi terakhir setiap pelanggan. Pelanggan yang baru saja bertransaksi mendapatkan skor lebih tinggi.
     - **Frequency**: Hitung jumlah transaksi yang dilakukan oleh setiap pelanggan dalam periode waktu tertentu. Pelanggan dengan lebih banyak transaksi mendapatkan skor lebih tinggi.
     - **Monetary**: Hitung total pengeluaran setiap pelanggan. Pelanggan yang menghabiskan lebih banyak uang mendapatkan skor lebih tinggi.

4. **Segmentasi Pelanggan**:
   - Berdasarkan skor RFM, segmentasikan pelanggan ke dalam kelompok yang berbeda. Misalnya, pelanggan dengan skor tinggi di semua kategori dapat dianggap sebagai pelanggan VIP, sementara mereka dengan skor rendah mungkin dianggap sebagai pelanggan yang perlu diaktifkan kembali.

5. **Analisis dan Tindakan**:
   - Analisis setiap segmen untuk memahami karakteristik dan perilaku pelanggan. Gunakan informasi ini untuk mengembangkan strategi pemasaran yang sesuai. Misalnya, tawarkan diskon atau promosi khusus kepada pelanggan VIP untuk meningkatkan loyalitas mereka, atau kirimkan kampanye reaktivasi kepada pelanggan yang tidak aktif.

6. **Evaluasi dan Penyesuaian**:
   - Evaluasi efektivitas strategi yang diterapkan dengan memantau metrik kinerja seperti peningkatan penjualan atau tingkat retensi pelanggan. Berdasarkan hasil evaluasi, lakukan penyesuaian pada strategi pemasaran untuk meningkatkan hasil di masa depan.

Dengan menggabungkan RFM Analysis dengan teknik data mining seperti clustering, klasifikasi, dan association rule mining dapat memberikan wawasan yang lebih mendalam dan strategi pemasaran yang lebih efektif. Berikut adalah tujuan dari masing-masing langkah tersebut:

1. **Clustering**: 
   - Clustering digunakan untuk mengelompokkan pelanggan dengan nilai RFM yang serupa. Ini membantu dalam mengidentifikasi segmen pelanggan yang berbeda, sehingga memungkinkan penerapan strategi pemasaran yang berbeda untuk setiap segmen. Misalnya, pelanggan dengan nilai Recency dan Frequency tinggi mungkin memerlukan strategi yang berbeda dibandingkan dengan pelanggan dengan nilai Monetary tinggi tetapi Recency rendah.

2. **Klasifikasi**:
   - Setelah pelanggan dikelompokkan, klasifikasi dapat digunakan untuk memprediksi perilaku pelanggan di masa depan berdasarkan variabel demografis dan nilai RFM mereka. Ini membantu dalam menargetkan profil pelanggan dengan lebih jelas dan mengantisipasi kebutuhan mereka.

3. **Association Rule Mining**:
   - Association rule mining digunakan untuk menemukan pola pembelian yang sering terjadi di antara segmen pelanggan. Ini membantu dalam memberikan rekomendasi produk yang lebih baik dan memenuhi kebutuhan pelanggan dengan lebih efektif. Misalnya, jika pelanggan dalam segmen tertentu sering membeli produk A dan B bersama-sama, maka produk tersebut dapat direkomendasikan kepada pelanggan lain dalam segmen yang sama.

Menggunakan ketiga teknik ini secara bersamaan dapat memberikan intelijen pasar yang lebih baik dan membantu manajer pemasaran dalam mengembangkan strategi yang lebih baik.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/kaggle/input/rfmdata/dataRFM.csv',encoding='ISO-8859-1',sep=',')
df
```

```python
df = df[df.Country == 'United Kingdom']
df
```

```python
df['InvoiceDate'].max()
```

```python
import datetime as dt
now = dt.date(2011,12,9)
df['date'] = pd.DatetimeIndex(df.InvoiceDate).date
df
```

```python
recency_df = df.groupby(['CustomerID'],as_index=False)['date'].max()
recency_df.columns = ['CustomerID','LastPurchaseDate']
recency_df.head()
```

```python
recency_df['Recency'] = recency_df.LastPurchaseDate.apply(lambda x : (now - x).days)
recency_df.drop(columns=['LastPurchaseDate'],inplace=True)
recency_df
```

```python
frequency_df = df.copy()
frequency_df.drop_duplicates(subset=['CustomerID','InvoiceNo'], keep="first", inplace=True) 
frequency_df = frequency_df.groupby('CustomerID',as_index=False)['InvoiceNo'].count()
frequency_df.columns = ['CustomerID','Frequency']
frequency_df.head()
```

```python
df['Total_cost'] = df['UnitPrice'] * df['Quantity']
monetary_df=df.groupby('CustomerID',as_index=False)['Total_cost'].sum()
monetary_df.columns = ['CustomerID','Monetary']
monetary_df.head()
```

```python
rf = recency_df.merge(frequency_df,left_on='CustomerID',right_on='CustomerID')

#combibe rf frame with monetary values..

rfm = rf.merge(monetary_df,left_on='CustomerID',right_on='CustomerID')

rfm.set_index('CustomerID',inplace=True)
rfm.head()
```

```python
rfm_segmentation = rfm.copy()
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, d = quartiles dict)

#lower the recency, good for store..
def RScore(x,p,d):
    if x <= d[p][0.2]:
        return 1
    elif x <= d[p][0.4]:
        return 2
    elif x <= d[p][0.6]: 
        return 3
    elif x <= d[p][0.8]: 
        return 4
    else:
        return 5
    
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)

#higher value of frequency and monetary lead to a good consumer. here higher value = 1 in reverse way.

def FMScore(x,p,d):
    if x <= d[p][0.2]:
        return 5
    elif x <= d[p][0.4]:
        return 4
    elif x <= d[p][0.6]: 
        return 3
    elif x <= d[p][0.8]: 
        return 2
    else:
        return 1
```

```python
quantile = rfm.quantile(q=[0.2,0.4,0.6,0.8])
quantile
```

```python
rfm_segmentation['R_Quartile'] = rfm_segmentation['Recency'].apply(RScore,args=('Recency',quantile))
rfm_segmentation['F_Quartile'] = rfm_segmentation['Frequency'].apply(FMScore, args=('Frequency',quantile))
rfm_segmentation['M_Quartile'] = rfm_segmentation['Monetary'].apply(FMScore, args=('Monetary',quantile))
```

```python
quantile.to_dict()
```

```python
rfm_segmentation['RFMScore'] = rfm_segmentation.R_Quartile.map(str) \
                            + rfm_segmentation.F_Quartile.map(str) \
                            + rfm_segmentation.M_Quartile.map(str)
rfm_segmentation.head()
```

```python
rfm_segmentation['Total_score'] = rfm_segmentation['R_Quartile'] + rfm_segmentation['F_Quartile'] + \
rfm_segmentation['M_Quartile']
```
